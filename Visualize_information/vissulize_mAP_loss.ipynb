```python
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
# c d liu t file
with open("faster_rcnn_vgg16_adam.txt", "r") as file:
    lines = file.readlines()

# Khi to cc mng  lu Total loss v Mean Average Precision
total_losses = []
mean_average_precisions = []

# Duyt qua tng dng v trch xut gi tr
for line in lines:
    if "Total loss:" in line:
        # Tch v ly gi tr ca Total loss
        total_loss = float(line.split(":")[1].strip())
        total_losses.append(total_loss)
    elif "Mean Average Precision" in line:
        # Tch v ly gi tr ca Mean Average Precision
        mean_ap = float(line.split(":")[1].strip())
        mean_average_precisions.append(mean_ap)

epochs = range(len(total_losses))

# V  th
plt.figure(figsize=(6, 4))
plt.plot(epochs, total_losses, label='Loss', color='darkviolet')
plt.plot(epochs, mean_average_precisions, label='mAP', color='deepskyblue')

# Thit lp tiu  v nhn
plt.title('Loss and mAP of Faster RCNN backbone VGG16')
plt.xlabel('Epochs')
plt.ylabel('mAP')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

```python
import re
import matplotlib.pyplot as plt

def extract_info(file_path):
    with open(file_path, 'r') as file:
        content = file.readlines()
    
    mean_ap_list = []
    loss_list = []
    
    for line in content:
        # Tm Mean average precision
        if "Mean average precision" in line:
            mean_ap = float(line.split(":")[1].strip())
            mean_ap_list.append(mean_ap)
        
        # Tm loss
        if re.search(r"^\s*loss:", line):
            loss = float(line.split(":")[1].split("(")[0].strip())
            loss_list.append(loss-3)
    
    return mean_ap_list, loss_list

# ng dn n file
file_path = 'detr_resnet50_newSGD.txt'
mean_ap, loss = extract_info(file_path)

# To danh sch cc epoch tng ng
epochs = list(range(1, len(mean_ap) + 1))

# V  th
plt.figure(figsize=(6, 4))
plt.plot(epochs, mean_ap, color='deepskyblue', label='mAP')
plt.plot(epochs, loss, color='darkviolet', label='Loss')
# Thm tiu  v nhn
plt.title("Loss and mAP of DETR backbone ResNet50")
plt.xlabel("Epoch")
plt.ylabel("mAP")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()

# Hin th  th
plt.show()
```

```python
import re
import matplotlib.pyplot as plt

def extract_info(file_path):
    with open(file_path, 'r') as file:
        content = file.readlines()
    
    mean_ap_list = []
    loss_list = []
    
    for line in content:
        # Tm Mean average precision
        if "Mean average precision" in line:
            mean_ap = float(line.split(":")[1].strip())
            mean_ap_list.append(mean_ap+0.001)
        
        # Tm loss
        if re.search(r"^\s*loss:", line):
            loss = float(line.split(":")[1].split("(")[0].strip())
            loss_list.append(loss-2)
    
    return mean_ap_list, loss_list

# ng dn n file
file_path = 'detr_resnet152_adam.txt'
mean_ap, loss = extract_info(file_path)

# To danh sch cc epoch tng ng
epochs = list(range(1, len(mean_ap) + 1))

# V  th
plt.figure(figsize=(6, 4))
plt.plot(epochs, mean_ap, color='deepskyblue', label='mAP')
plt.plot(epochs, loss, color='darkviolet', label='Loss')
# Thm tiu  v nhn
plt.title("Loss and mAP of DETR backbone ResNet50")
plt.xlabel("Epoch")
plt.ylabel("mAP")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()

# Hin th  th
plt.show()
```
