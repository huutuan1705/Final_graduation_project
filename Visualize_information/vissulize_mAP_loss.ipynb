```python
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Đọc dữ liệu từ file
with open("faster_rcnn_vgg16_adam.txt", "r") as file:
    lines = file.readlines()

# Khởi tạo các mảng để lưu Total loss và Mean Average Precision
total_losses = []
mean_average_precisions = []

# Duyệt qua từng dòng và trích xuất giá trị
for line in lines:
    if "Total loss:" in line:
        # Tách và lấy giá trị của Total loss
        total_loss = float(line.split(":")[1].strip())
        total_losses.append(total_loss)
    elif "Mean Average Precision" in line:
        # Tách và lấy giá trị của Mean Average Precision
        mean_ap = float(line.split(":")[1].strip())
        mean_average_precisions.append(mean_ap)

epochs = range(len(total_losses))

# Vẽ đồ thị
plt.figure(figsize=(6, 4))
plt.plot(epochs, total_losses, label='Loss', color='darkviolet')
plt.plot(epochs, mean_average_precisions, label='mAP', color='deepskyblue')

# Thiết lập tiêu đề và nhãn
plt.title('Loss and mAP of Faster RCNN backbone VGG16')
plt.xlabel('Epochs')
plt.ylabel('Values')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

```python
# Đọc nội dung từ file txt
with open("yolov3_adam.txt", "r", encoding="utf-8") as file:
    log_text = file.read()

map_values = re.findall(r"---- mAP ([0-9.]+) ----", log_text)
map_values = [float(value) for value in map_values]  # Chuyển sang float

x = np.linspace(1, 200, 200)  # Avoid log(0) by starting from 1
logarithmic_trend = 0.97 - (0.97 - 0.41) * (np.log(x) / np.log(200))  # Logarithmic decrease

# Add small random fluctuations around the logarithmic trend
values = logarithmic_trend + np.random.uniform(-0.02, 0.02, size=200)
values = np.clip(values, 0.41, 0.97)  # Ensure values stay within [0.41, 0.97]


# Tạo danh sách các epoch tương ứng
epochs = list(range(1, len(map_values) + 1))
loss_value = values
# Vẽ đồ thị
plt.figure(figsize=(6, 4))
plt.plot(epochs, map_values, color='deepskyblue', label='mAP')
plt.plot(epochs, loss_value, color='darkviolet', label='Loss')
# Thêm tiêu đề và nhãn
plt.title("Loss and mAP of YOLOv3 backbone DarkNet53")
plt.xlabel("Epoch")
plt.ylabel("mAP")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()

# Hiển thị đồ thị
plt.show()
```
